```python
#Import the package
import spacy
import pandas as pd
```

```python
nlp = spacy.load("en_core_web_sm")   #Reference- https://spacy.io/models
```

## Tokenization

  Splitting a piece of text into words, symbol,punctuation and other elements

### Word Token

```python
doc1=nlp("As a DataScientist ifeel it's the most happiest job i've ever had")
doc1
```

```python
for x,token in enumerate(doc1):
  print(x,token.text)
```

### Sentence Tokenization

```python
sent_doc='Machine learning is an application of artificial intelligence (AI) that provides systems the ability to automatically learn and improve from experience without being explicitly programmed. Machine learning focuses on the development of computer programs that can access data and use it learn for themselves.The process of learning begins with observations or data, such as examples, direct experience, or instruction, in order to look for patterns in data and make better decisions in the future based on the examples that we provide. The primary aim is to allow the computers learn automatically without human intervention or assistance and adjust actions accordingly.'
sent_doc
```

```python
sentence_doc1=nlp(sent_doc)
```

```python
for x,i in enumerate(sentence_doc1.sents):
  print(x,i)
```

### Lemmatization
    
    Lemmatization is the process of reducing a word to its base form

```python
doc_lemma=nlp('Practice,practiced,practicing')
doc_lemma
```

```python
for token in doc_lemma:
  print(token.text,token.lemma_,token.lemma_.lower().strip())
```

```python
doc_lemma1=nlp('He was crying incessantly in order to douse the flame of the pain caused to him. Charles cried because, he lost his mother in a horrible accident')
```

```python
#Method using dictonary  
#Ref: https://spacy.io/api/annotation
for token in doc_lemma1:
  print(f'{token.text:{20}} {token.lemma_:{20}} {token.pos_:{10}}')
```

```python
#Using Pandas
lemma=[]
for x,token in enumerate(doc_lemma):
  lemma.append([x,token.text,token.lemma_,token.pos_])
pd.DataFrame(lemma,columns=['Index','Token','Lemma','pos'])
```
